# Data Cleaning for Descriptive Statistics

This notebook performs minimal and intentional data cleaning required
to compute descriptive statistics correctly.

The goal is not to fully sanitize the dataset, but to prepare key columns
for meaningful statistical analysis.

```python
import pandas as pd
import numpy as np

df = pd.read_csv("../data/startup_funding.csv")
```

```python
df.shape
df.columns
```

```python
df["Amount in USD"].head(15)
```

```python
funding_numeric = (
    df["Amount in USD"]
    .astype(str)
    .str.replace(",", "", regex=True)
)
```

```python
funding_numeric = pd.to_numeric(
    funding_numeric,
    errors="coerce"
)
```

```python
funding_numeric.head()
```

```python
funding_numeric.isnull().sum()
```

```python
funding_numeric.shape
```

```python
df_funding = df.loc[funding_numeric.notna()].copy()
df_funding["Amount in USD"] = funding_numeric[funding_numeric.notna()]
```
